```python
# https://www.kaggle.com/code/astrung/lstm-model-with-item-infor-fix-missing-last-item
def add_last_item(old_interaction, last_item_id, max_len=50):
    new_seq_items = old_interaction['item_list'][-1]
    if old_interaction['item_length'][-1].item() < max_len:
        new_seq_items[old_interaction['item_length'][-1].item()] = last_item_id
    else:
        new_seq_items = torch.roll(new_seq_items, -1)
        new_seq_items[-1] = last_item_id
    return new_seq_items.view(1, len(new_seq_items))

def predict_for_all_item(user_id, dataset, model, config):
    model.eval()
    with torch.no_grad():
        index = np.isin(dataset3.inter_feat[dataset.uid_field].numpy(), user_id)
        input_interaction = dataset[index]
        test = {
            'item_list': add_last_item(input_interaction, 
                                          input_interaction['item'][-1].item(), model.max_seq_length),
            'item_length': torch.tensor(
                [input_interaction['item_length'][-1].item() + 1
                 if input_interaction['item_length'][-1].item() < model.max_seq_length else model.max_seq_length])
        }
        new_inter = Interaction(test)
        new_inter = new_inter.to(config['device'])
        new_scores = model.full_sort_predict(new_inter)
        new_scores = new_scores.view(-1, 6808)
        new_scores[:, 0] = -np.inf
    return new_scores, new_inter
```

```python
from tqdm import tqdm
```

```python
inference3 = torch.zeros(1, 6808).to('cuda:0')

for i in tqdm(range(1, 31361)):
    inference3 = torch.cat((inference3, predict_for_all_item(i, dataset3, model3, config3)[0]), 0)
```
